```python
import numpy as np

# ---------- FUNCTIONS ----------
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

def conv2d(image, kernel):
    k = kernel.shape[0]
    output_size = image.shape[0] - k + 1
    result = np.zeros((output_size, output_size))

    for i in range(output_size):
        for j in range(output_size):
            region = image[i:i+k, j:j+k]
            result[i,j] = np.sum(region * kernel)
    return result

# ---------- INPUT ----------
image = np.array([
    [1,2,3,0,1],
    [0,1,2,3,1],
    [1,2,1,0,0],
    [0,1,2,1,2],
    [1,0,1,2,1]
])

target = np.array([[1]])

# ---------- INITIAL WEIGHTS ----------
np.random.seed(1)
kernel = np.random.randn(3,3)
fc_weights = np.random.randn(9,1)

print("Initial Kernel:\n", kernel)
print("\nInitial FC Weights:\n", fc_weights)

# ---------- BEFORE TRAINING ----------
conv_out = conv2d(image, kernel)
activated = relu(conv_out)
flatten = activated.reshape(1,-1)
output = flatten.dot(fc_weights)

print("\nOutput Before Training:\n", output)

loss = np.mean((target - output)**2)
print("\nInitial Loss:", loss)

# ---------- TRAINING ----------
lr = 0.01
epochs = 200

for epoch in range(epochs):

    # Forward
    conv_out = conv2d(image, kernel)
    activated = relu(conv_out)
    flatten = activated.reshape(1,-1)
    output = flatten.dot(fc_weights)

    # Loss
    error = target - output

    # Backprop
    d_output = error
    d_fc = flatten.T.dot(d_output)

    d_flat = d_output.dot(fc_weights.T)
    d_act = d_flat.reshape(activated.shape) * relu_derivative(activated)

    # Update
    fc_weights += lr * d_fc
    kernel += lr * np.ones_like(kernel) * np.mean(d_act)

# ---------- AFTER TRAINING ----------
print("\nUpdated Kernel:\n", kernel)
print("\nUpdated FC Weights:\n", fc_weights)

conv_out = conv2d(image, kernel)
activated = relu(conv_out)
flatten = activated.reshape(1,-1)
final_output = flatten.dot(fc_weights)

print("\nFinal Output:\n", final_output)

final_loss = np.mean((target - final_output)**2)
print("\nFinal Loss:", final_loss)
```
