```python
#RNN Backpropagation

import numpy as np

# ---------- FUNCTIONS ----------
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - x**2

# ---------- INPUT ----------
inputs = [0.5, 0.2, 0.1]
target = 0.3

# ---------- INITIAL WEIGHTS ----------
np.random.seed(2)
Wx = np.random.randn()
Wh = np.random.randn()
Wy = np.random.randn()

print("Initial Weights:")
print("Wx:", Wx)
print("Wh:", Wh)
print("Wy:", Wy)

# ---------- BEFORE TRAINING ----------
h = 0
states = []

for x in inputs:
    h = tanh(Wx*x + Wh*h)
    states.append(h)

output = Wy * h
print("\nOutput Before Training:", output)

loss = (target - output)**2
print("\nInitial Loss:", loss)

# ---------- TRAINING ----------
lr = 0.05
epochs = 200

for epoch in range(epochs):

    # Forward
    h = 0
    states = []
    for x in inputs:
        h = tanh(Wx*x + Wh*h)
        states.append(h)

    output = Wy * h
    error = target - output

    # Backprop
    dWy = error * h
    dh = error * Wy

    dWx, dWh = 0, 0

    for t in reversed(range(len(inputs))):
        dh_raw = dh * tanh_derivative(states[t])
        dWx += dh_raw * inputs[t]
        dWh += dh_raw * (states[t-1] if t > 0 else 0)
        dh = dh_raw * Wh

    # Update
    Wx += lr * dWx
    Wh += lr * dWh
    Wy += lr * dWy

# ---------- AFTER TRAINING ----------
print("\nUpdated Weights:")
print("Wx:", Wx)
print("Wh:", Wh)
print("Wy:", Wy)

# Final output
h = 0
for x in inputs:
    h = tanh(Wx*x + Wh*h)

final_output = Wy * h
print("\nFinal Output:", final_output)

final_loss = (target - final_output)**2
print("\nFinal Loss:", final_loss)
```
