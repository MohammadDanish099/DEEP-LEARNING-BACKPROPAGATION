```python
#ANN Backpropagation
import numpy as np

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_grad(x):
    return x * (1 - x)

# Input (XOR problem)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])

# Network architecture
input_nodes = 2
hidden_nodes = 3
output_nodes = 1

# Initialize weights (different style)
np.random.seed(42)
W1 = np.random.uniform(-1, 1, (input_nodes, hidden_nodes))
W2 = np.random.uniform(-1, 1, (hidden_nodes, output_nodes))

lr = 0.5
epochs = 5000

for i in range(epochs):

    # Forward pass
    hidden_input = np.dot(X, W1)
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2)
    final_output = sigmoid(final_input)

    # Error
    error = Y - final_output

    # Backpropagation
    d_output = error * sigmoid_grad(final_output)

    hidden_error = d_output.dot(W2.T)
    d_hidden = hidden_error * sigmoid_grad(hidden_output)

    # Weight updates
    W2 += hidden_output.T.dot(d_output) * lr
    W1 += X.T.dot(d_hidden) * lr

# Output
print("ANN Output:\n", final_output.round(3))
```
